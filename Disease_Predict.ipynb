## Data Processing

```python
# Convert target variable to numerical (0 and 1)
dataset['Heart Disease'] = dataset['Heart Disease'].map({'Presence': 1, 'Absence': 0})
```

```python
# Check if conversion worked
print(dataset['Heart Disease'].value_counts())
```

## Prepare data for modeling

```python
from sklearn.model_selection import train_test_split

# Separate features (X) and target (y)
X = dataset.drop('Heart Disease', axis=1)
y = dataset['Heart Disease']
```

```python
# Split into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(f"Training set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

## Building classification models

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=5000)
log_reg.fit(X_train, y_train)
```

## Making Predictions

```python
predictions = log_reg.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions))
```
